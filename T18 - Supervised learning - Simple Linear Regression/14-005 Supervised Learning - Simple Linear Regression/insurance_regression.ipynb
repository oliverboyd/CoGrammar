```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
insurance_df = pd.read_csv("insurance.csv")
```

```python
sns.scatterplot(data=insurance_df, x='age', y='charges', color='b')
plt.xlabel('Age')
plt.ylabel('Charges')
```

```python
X = insurance_df['age'].values.reshape(-1, 1) # create a 2D array
y = insurance_df['charges'].values
```

```python
simple_model = LinearRegression()
simple_model.fit(X, y)
y_pred = simple_model.predict(X)
```

```python
sns.scatterplot(data=insurance_df, x='age', y='charges', color='b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.plot(X, y_pred, color='r')
```
